```python
# Seleccionamos el sub_anndata_muchari_budinger donde study == "Mishari_Budinger_2018"
import scanpy as sc
import celltypist
from celltypist import models

sub_anndata_muchari_budinger = sc.read_h5ad('/home/mdiaz/scRNA-seq_LungProject/atlas_reference/data/processed/SinAnotar_muchari_budinger.h5ad')
```

```python
sub_anndata_muchari_budinger.var_names[:10].tolist()
```

```python
import mygene

mg = mygene.MyGeneInfo()
query = mg.querymany(sub_anndata_muchari_budinger.var_names.tolist(), scopes='ensembl.gene', fields='symbol', species='human')

# Crear un diccionario de mapeo
symbol_map = {entry['query']: entry.get('symbol', None) for entry in query if 'symbol' in entry}

# Filtrar genes con s√≠mbolo v√°lido
valid_genes = [g for g in sub_anndata_muchari_budinger.var_names if g in symbol_map]

# Subset y renombrar
adata_mapped = sub_anndata_muchari_budinger[:, valid_genes].copy()
adata_mapped.var_names = [symbol_map[g] for g in valid_genes]
```

```python
sc.pp.normalize_total(adata_mapped, target_sum=1e4)
sc.pp.log1p(adata_mapped)
```

```python
models.models_description()
```

```python
healthy_model = models.Model.load(model = 'Human_Lung_Atlas.pkl')
```

```python
ipf_model = models.Model.load(model = 'Human_IPF_Lung.pkl')
```

```python
# Anotar usando los modelos 

# Anotaci√≥n con el modelo Human_Lung_Atlas
pred_atlas = celltypist.annotate(adata_mapped, model= healthy_model, majority_voting=True, mode='best match')

# Anotaci√≥n con el modelo Human_PF_Lung
pred_ipf = celltypist.annotate(adata_mapped, model= ipf_model, majority_voting=True, mode='best match')
```

```python
pred_atlas.predicted_labels

```

```python
adata_atlas = pred_atlas.to_adata()
adata_ipf = pred_ipf.to_adata()

```

```python
adata_atlas.obs_names.equals(adata_ipf.obs_names)
```

```python
# Asegurarnos de que los √≠ndices de las celdas coinciden
assert adata_atlas.obs_names.equals(adata_ipf.obs_names), "Los √≠ndices de c√©lulas no coinciden"

# Extraer labels y scores
df_atlas = adata_atlas.obs[['predicted_labels', 'conf_score']].rename(
    columns={'predicted_labels': 'label_atlas', 'conf_score': 'conf_atlas'}
)
df_ipf = adata_ipf.obs[['predicted_labels', 'conf_score']].rename(
    columns={'predicted_labels': 'label_ipf', 'conf_score': 'conf_ipf'}
)

# Combinar ambos DataFrames
df_combined = pd.concat([df_atlas, df_ipf], axis=1)

# Elegir la predicci√≥n con mayor confianza por celda
df_combined['final_label'] = df_combined.apply(
    lambda row: row['label_atlas'] if row['conf_atlas'] >= row['conf_ipf'] else row['label_ipf'], axis=1
)
df_combined['final_conf'] = df_combined[['conf_atlas', 'conf_ipf']].max(axis=1)

# O si prefieres una versi√≥n final como Series para agregar al AnnData:
adata_atlas.obs['final_label'] = df_combined['final_label']
adata_atlas.obs['final_conf'] = df_combined['final_conf']
```
